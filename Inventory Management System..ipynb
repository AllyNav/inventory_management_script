# Product Details

1. Product ID
2. Name of the Product
3. Price of Produt
4. Quantity

```python
import time

#Reading my Inventory 
fd = open('Inventory.txt','r')
products = fd.read().split('\n')
fd.close()

#Taking input for user what it wants to purchase
ui_username = input("Enter your Name:")
ui_phone= input("Enter your phone number:")
ui_email= input("Enter your email id:")
ui_prod_id = input("Enter product ID: ")
ui_prod_qn = input("Enter product Quantity: ")

updated_product_lst = []

for product in products:
    
    prod_details = product.split(',')
    
    if(prod_details[0] == ui_prod_id):
        #enough product is present and reducing it from inventory
        if(int(ui_prod_qn) <= int(prod_details[3])):
           
            print("-----------------------------")
            print("Product Name     : ", prod_details[1])
            print("Price            : ", prod_details[2]) 
            print("Quantity         : ", ui_prod_qn) 
            print("-----------------------------")
            print("Billing Amount   : ", int(ui_prod_qn) * int(prod_details[2]))
            print("-----------------------------")
            
            #updating our inventory list
            prod_details[3] = str(int(prod_details[3]) - int(ui_prod_qn))
            
            #Generating  Sales in Sales.txt
            fd= open("Sales.txt", 'a')
            sales_detail = ui_username+","+ui_phone+","+ui_email+","+ui_prod_id+","+ui_prod_qn+","+str(int(ui_prod_qn) * int(prod_details[2]))+","+time.ctime()+"\n"
            fd.write(sales_detail)
            fd.close()
            
        else:
            #if enough quantity is not present
           
            print("Sorry, we are not having enough quantity.")
            print("We are having only ", prod_details[3]," quantity. ")
            print("Would you like to purcchase it? Happy to Serve")
            ch = input("Press Y/N ")
                
            if (ch=='Y' or ch== 'y'):
                #if product is not enough and wants to buy remaing quantity
                print("-----------------------------")
                print("Product Name     : ", prod_details[1])
                print("Price            : ", prod_details[2]) 
                print("Quantity         : ", ui_prod_qn) 
                print("-----------------------------")
                print("Billing Amount   : ", int(prod_details[3]) * int(prod_details[2]))
                print("-----------------------------")

                prod_details[3] = '0'
            
            else:
                print("Thanks")
             
           
                    
        
    updated_product_lst.append(prod_details)
    

    
lst = []

#Updating my inventory String
for i in updated_product_lst:
    prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
    lst.append(prod)

lst[-1] = lst[-1][:-1]
    
fd = open('Inventory.txt','w')

#Updating my inventory file
for i in lst:
    fd.write(i)

fd.close()

print("-----------------------------")
print("Inventory Updated")
```
